```python
import numpy as np
from heapq import heappop, heappush
from interface import BlackBox
import matplotlib.pyplot as plt

hyperpar = 2
def mymse(x, y):
    return np.square(x - y)
    #return (x - y) ** 2

def add_list(bb, diff, x_mod, j, sc):
    x_modd = np.copy(x_mod)
    x_modd[j] = min(255, x_mod[j] + 1)
    scor = bb.calc(x_modd)
    if scor - sc > 0:
        mmm = (mymse(x_modd[j], x_orig[j])-mymse(x_mod[j], x_orig[j]))
        mmm = 1
        if mmm < 0:
            print("YEAS")
            #mmm = 0.000001
        heappush(diff, ((-(scor-sc)**(1/hyperpar))/mmm, j, 1))
        return
    else:
        x_modd[j] = max(0, x_mod[j] - 1)
        scor = bb.calc(x_modd)
        if scor - sc > 0:
            mmm = (mymse(x_modd[j], x_orig[j])-mymse(x_mod[j], x_orig[j]))
            mmm = 1
            if mmm < 0:
                print("YEAS")
                #mmm = 0.000001
            heappush(diff, ((-(scor-sc)**(1/hyperpar))/mmm, j, -1))
            return

bb = BlackBox()

def mse(x, y):
    return np.square(x - y).mean()
    #return sum((x-y)**2) / len(x)

def add_random(x):
    noise = np.random.randint(0, x.shape[0], 1)
    y = x.copy()
    y[noise] = 0
    return y
```
